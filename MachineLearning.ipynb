```python
from datasets import load_dataset

dataset = load_dataset("uitnlp/vietnamese_students_feedback")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score,classification_report

train_data = dataset['train']
test_data = dataset['test']
val_data = dataset['validation']

# Split the dataset into training and testing sets
train_texts = train_data['sentence']
train_labels = train_data['sentiment']

test_texts = test_data['sentence']
test_labels = test_data['sentiment']
# Convert text data to TF-IDF features
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train_texts)
X_test = vectorizer.transform(test_texts)

# Train SVM
svm_classifier = MLPClassifier(hidden_layer_sizes=(200,),random_state=1, max_iter=500)
svm_classifier.fit(X_train, train_labels)

# Make predictions on the test set
predictions = svm_classifier.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(test_labels, predictions)
f1_score = f1_score(test_labels, predictions, average="macro")
print(f"Accuracy: {accuracy}")
print(f"F1_score: {f1_score}")
```

```python
print(classification_report(test_labels,predictions))
```
